```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Medicaldataset.csv')
df
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue', stat="count")
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Heart rate'], bins=20, kde=True, color='salmon',stat='count')
plt.title('Heart Rate Distribution')
plt.xlabel('Heart Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Blood sugar'], bins=20, kde=True, color='green',stat='count')
plt.title('Blood Sugar Distribution')
plt.xlabel('Blood Sugar')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Troponin'], bins=20, kde=True, color='red',stat='count')
plt.title('Troponin Distribution')
plt.xlabel('Troponin')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(x='Result', data=df, palette='pastel')
plt.title('Result Distribution')
plt.xlabel('Result')
plt.ylabel('Count')
plt.show()
```
